***Outliers Step 2: Statistical Analysis:***

```python
# Determine IQR
Q1 = df['Median'].quantile(0.25)
Q3 = df['Median'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers outside of Q1 - 1.5*IQR and Q3 + 1.5*IQR
outliers = df[(df['Median'] < (Q1 - 1.5 * IQR)) | (df['Median'] > (Q3 + 1.5 * IQR))]

print("Number of outliers detected in ‘Median’ salary:", len(outliers))
```

***Outliers Step 3: Decision on Outliers:***

The output from the code's print statement for the number of outliers detected in the `Median` salary is:

- Number of outliers detected in 'Median' salary: 6

This indicates that there are 6 majors with median salaries significantly higher or lower than the majority, qualifying them as outliers based on the statistical analysis using the Interquartile Range method.

Keeping the outliers in the `Median` salary analysis is essential because they represent specialized fields with high earning potential, accurately reflecting real-world salary variations. These outliers provide crucial insights for prospective students making career decisions and ensure a comprehensive analysis of the job market for recent graduates. By including these data points, the analysis maintains its integrity, offering a nuanced understanding of employment outcomes across various majors and highlighting the exceptional earning opportunities in certain fields. Removing these outliers would not only skew the dataset towards lower-earning majors but also omit critical information necessary for informed decision-making and policy formulation.

## Exploratory Data Analysis

***Major Distribution:***

```python
major_counts = df["Major_category"].value_counts()

plt.figure(figsize=(10, 6))
major_counts.plot(kind='barh')
plt.xlabel('Number of Students')
plt.ylabel('Major')
plt.title('Number of Students in Each Major')
plt.show();
```

The chart analyzed shows the popularity of various majors by depicting the number of students in each field. Engineering, education, and humanities & liberal arts majors emerge as the most chosen majors, with engineering leading by a significant margin. This visual distribution indicates the potential saturation in the job market for these majors, which could impact graduates’ wages and employment opportunities. The chart serves as a tool to understand trends in education and as a guide for students to gauge the competitive landscape of their chosen fields. 

***Major & Percentage of Unemployed Individuals:***

```python
major_grouped = df.groupby("Major_category").mean().reset_index()
major = major_grouped["Major_category"]
unemployment_rate = sorted(major_grouped["Unemployment_rate"])

plt.figure(figsize=(10, 6))
plt.barh(major, unemployment_rate, color='skyblue')
plt.xlabel('Percentage of Unemployed Individuals (%)')
plt.ylabel('Major')
plt.title('Major vs Percentage of Unemployed Individuals')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest value at the top
plt.show();
```

The bar chart represents the unemployment rate by major category, with social science, psychology & social work, and physical sciences having higher rates, which could suggest a competitive job market or limited openings in these fields. In contrast, education and engineering show notably lower unemployment rates, indicating these sectors might offer more robust job opportunities. This visualization aids students and educators in understanding the employment landscape and could be critical for informing decisions about educational and career pathways. 

***Major and Wage Median:***


```python
import matplotlib.pyplot as plt

major_grouped = df.groupby("Major_category").mean().reset_index()
sorted_major_grouped = major_grouped.sort_values(by='Median', ascending=False)

major = sorted_major_grouped["Major_category"]
median_wage = sorted_major_grouped["Median"]

plt.figure(figsize=(10, 6))
plt.barh(major, median_wage, color='skyblue')
plt.xlabel('Median Wage')
plt.ylabel('Major')
plt.title('Major vs Median Wage')
plt.show();
```

The provided bar chart categorizes academic majors by median wage, revealing that engineering, business, and computer and mathematics majors are at the top end of the wage spectrum. In contrast, the median wages for psychology & social work, humanities & liberal arts, and education are at the lower end. This suggests that technical and commercially-oriented majors may offer better financial returns in the early stages of a career. Such insights are crucial for students considering the long-term economic benefits of their chosen field of study. 


***Impact of Major Popularity and Full-time Employment Proportion on Early Career Wages (Multivariable Relationships)***

```python
#Calculate the proportion of full-time employment to define bins and create the new categorical variable
df['Full_time_prop'] = df['Full_time'] / df['Total']

#Define bins for the proportion of full-time employment
bins = [0, 0.33, 0.66, 1]
labels = ['Low (<33%)', 'Medium (33%-66%)', 'High (>66%)']
df['Full_time_prop_category'] = pd.cut(df['Full_time_prop'], bins=bins, labels=labels, include_lowest=True)

# Scatter plot with categorical colors
plt.figure(figsize=(12, 6))
scatter = sns.scatterplot(data=df, x='Total', y='Median', hue='Full_time_prop_category',
                          style='Full_time_prop_category', palette='viridis', s=100, alpha=0.7)

plt.title('Median Wage vs. Total Graduates, Categorized by Full-time Employment Proportion')
plt.xlabel('Total Graduates (Log Scale)')
plt.ylabel('Median Early Career Wage ($)')
plt.xscale('log')  # Using a log scale for Total Graduates for better visibility

plt.legend(title='Proportion of Full-time Employment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, which="both", ls="--")
plt.tight_layout()
plt.show();
```

This visualization explores the relationship between the popularity of college majors, as measured by total graduates, and early career wages, while also considering the impact of the proportion of full-time employment among graduates. By categorizing the proportion of full-time employment into three distinct ranges—Low (<33%), Medium (33%-66%), and High (>66%)—and displaying this on a scatter plot against the logarithmic scale of total graduates, we gain insights into how these factors interact. The analysis reveals that the size of the major and the rate of full-time employment do not show a straightforward correlation with early career wages. Majors with varying levels of popularity and full-time employment proportions lead to a wide spectrum of median early career wages. This complexity underscores the multifaceted nature of the factors influencing early career outcomes, suggesting that the economic value of a college major cannot be determined by popularity or employment status alone. This finding contributes to our understanding of the original research question by highlighting the importance of considering multiple dimensions when evaluating how academic major choices influence early career wages and employment relevance.

***Relationships Between Variables:***


```python
# Select the relevant columns for scatter plots (18 variables)
each_variable = ['Rank', 'Major_code', 'Total', 'Men', 'Women', 
'ShareWomen', 'Sample_size', 'Employed', 'Full_time', 
'Part_time', 'Full_time_year_round', 'Unemployed', 
'Unemployment_rate', 'P25th', 'P75th', 'College_jobs', 
               	'Non_college_jobs', 'Low_wage_jobs']

# Set up subplots
fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(25, 25))
fig.tight_layout(pad=4.0)

# Create scatter plots
for i, col in enumerate(each_variable):
    row = i // 4
    col = i % 4
    axes[row, col].scatter(df[each_variable[i]], df['Median'], alpha=0.5)
    axes[row, col].set_title(f'Scatter Plot: {each_variable[i]} vs. Median Salary')
    axes[row, col].set_xlabel(each_variable[i])
    axes[row, col].set_ylabel('Median Salary')

plt.show();
```

Our exploratory data analysis has shed light on the interplay between academic majors and their associated earning potential. The bar charts and box plots underscore the trend that STEM fields typically offer high median salaries, a reflection of the market’s demand for these skills. The scatter plots further this narrative, revealing the inverse correlation between low-wage jobs and median salaries across majors. These visuals have given us a deeper understanding of our dataset, informing us that while certain majors may pave the way to higher salaries, the value of a major is multifaceted and can not be judged on earnings alone. For example, our bar chart illustrates that engineering majors not only dominate in terms of higher median salaries but also show a remarkable range within the field. This is indicated by the variability in the corresponding boxplot. Moreover, the scatter plots highlighted an interesting trend: majors with a high number of college-required jobs generally show higher median salaries. In contrast, those with non-college jobs/low-wage positions tend to have lower median salaries. In conclusion, this analysis forms a solid foundation for our next steps, guiding us in making data-driven inferences and providing actionable insights.  

Going further into the analysis of the scatter plots, if we look at the comparison between the total number of graduates to median salary shows no clear correlation, suggesting that the popularity of a major does not necessarily equate to higher earnings.
